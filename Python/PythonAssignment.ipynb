```python
def ispalindrome(val):
  for x in range(0,int(len(val)/2)):
    if val[x] == val[(x+1)-((x+1)*2)]:
      pass
    else:
      return 0
  return 1
```

```python
print(ispalindrome('12321'))
print(ispalindrome('124321'))
print(ispalindrome('malayalam'))
```

You will be given a sentence in the form of a string. You have to reverse the
order of the words in the sentence. Remember not to reverse the individual words, but
the order of words.

```python
def getReverse(str):
  wordlist=str.split(" ")
  newlist=[]
  for x in range(1,len(wordlist)+1):
    newlist.append(wordlist[-x])
  print(newlist)
    
```

```python
getReverse('You will be given a sentence')
```
